```python
import numpy as np 
import pandas as pd 
```

```python
movie = pd.read_csv('tmdb_5000_movies.csv')
credit = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movie.head()
```

```python
credit.head()
```

```python
movie = movie.merge(credit, on='title')
movie.head()
```

```python
movie = movie[['id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movie.head()
```

```python
import ast
```

```python
def convert(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
movie.isnull().sum()
```

```python
movie.dropna(inplace= True)
```

```python
movie.isnull().sum()
```

```python
movie['genres'] = movie['genres'].apply(convert)
movie.head()
```

```python
movie['keywords'] = movie['keywords'].apply(convert)
movie.head()
```

```python
def convert2(text):
    l = []
    c = 0
    for i in ast.literal_eval(text):
        if c!=3:
            l.append(i['character'])
            c+=1
        else:
            break
    return l
```

```python
movie['cast'] = movie['cast'].apply(convert2)
movie.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movie['crew'] = movie['crew'].apply(fetch_director)
movie.head()
```

```python
movie['overview']
```

```python
movie['overview'] = movie['overview'].apply(lambda x:x.split())
movie.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movie['cast'] = movie['cast'].apply(collapse)
movie['crew'] = movie['crew'].apply(collapse)
movie['genres'] = movie['genres'].apply(collapse)
movie['keywords'] = movie['keywords'].apply(collapse)
```

```python
movie.head()
```

```python
movie['tag'] = movie['overview'] + movie['genres'] + movie['keywords'] + movie['cast'] + movie['crew']
movie.head()
```

```python
new_movie = movie[['id', 'title', 'tag']]
new_movie.head()
```

```python
new_movie['tag'] = new_movie['tag'].apply(lambda x: " ".join(x))
new_movie.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_movie['tag']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
similarity
```

```python
new_movie[new_movie['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new_movie[new_movie['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_movie.iloc[i[0]].title)
```

```python
recommend('Batman')
```
